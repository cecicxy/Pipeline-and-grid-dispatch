```python
wind_solar=pd.read_csv('../data/renewable potential/wind_solar_potential.csv',index_col=0) #单位：吉千瓦时（e9KWh）
wind_solar
```

```python
wind_solar=wind_solar.iloc[:-1,:]
wind=wind_solar.loc[:,'W_Wind']
solar=wind_solar.loc[:,'W_PV']
sorted_wind = wind.sort_values(ascending=False)
sorted_solar = solar.sort_values(ascending=False)
```

```python
import matplotlib 
from matplotlib import cm
import matplotlib.ticker as ticker
fig, ax = plt.subplots(figsize=(7.083, 5), layout='constrained')
x=range(len(sorted_wind))  # 确定柱状图数量,可以认为是x方向刻度
y=sorted_wind  # y方向刻度

  #cm.get_cmap('viridis', len(wind_solar))  # 确定颜色
ax.set_xticks(range(len(sorted_wind)), sorted_wind.index,rotation='vertical',fontsize=10,fontname='Times New Roman')
plt.yticks(fontsize=10)
ax.bar(x, y,color='#a7e2b7')  # 绘制y刻度标签
ax.set_ylabel(r'$E_{wind}$  ($10^9$KWh)',fontsize=12,fontname='Times New Roman')  # y轴标签
# 在每个柱子上方添加数值
for i, v in enumerate(y):
    ax.text(i, v + 130, str(round(v,2)), ha='center', fontsize=10, fontname='Times New Roman',rotation='vertical')
    # 启用次要刻度线
ax.minorticks_on()
# 设置y轴的次要刻度线的位置
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())
# 禁用x轴的次要刻度线
ax.xaxis.set_minor_locator(ticker.NullLocator())
#设置网格刻度
plt.ylim(0, max(y) * 1.125)  # 设置y轴的最大值为最大y值的110%
plt.savefig('C:/Users/ASUS/Documents/research/hydrogen production/publication/Figure/renewable energy resource/bar_svg/wind_potential.svg')  # 保存为SVG格式
plt.show()
```

各省水电核电潜力

tracker的数据只统计>=75 MW的水电站，因此有些省份是0. 上海没有水力发电，天津在造，西藏和山东则分别于2004年、2007年取得零的突破，但截至目前体量仍然较小。

```python
hydropower_all=pd.read_excel("../data/renewable potential/Global-Hydropower-Tracker-May-2023.xlsx")
```

```python
hydropower_all=hydropower_all[(hydropower_all["Country 1"]=="China")& (hydropower_all["Binational"]=="No")]
```

```python
hydropower=hydropower_all[["Project Name","Project Name (local lang/script)","Status","Capacity (MW)","State/Province 1","Latitude","Longitude"]]
```

```python
hydropower=hydropower.reset_index(drop=True)
```

```python
hydropower[hydropower["State/Province 1"].isnull()==True]
```

```python
hydropower.loc[416,"State/Province 1"]="Yunnan"
hydropower.loc[777,"State/Province 1"]="Sichuan"
hydropower.loc[847,"State/Province 1"]="Hebei"
```

```python
hydropower[hydropower["State/Province 1"].isnull()==True]
```

```python
hydropower["Status"].unique()
```

```python
hydropower=hydropower[hydropower["Status"].isin(['operating','construction','pre-construction','announced'])]
```

```python
hydropower['State/Province 1'].unique() #名字不一样，需要统一
```

```python
hydropower['State/Province 1'].replace('Xinjiang Uygur Autonomous Region','Xinjiang',inplace=True)
hydropower['State/Province 1'].replace('Hunan Province','Hunan',inplace=True)
hydropower['State/Province 1'].replace('Tibet Autonomous Region','Tibet',inplace=True)
hydropower['State/Province 1'].replace('Ningxia Hui Autonomous Region','Ningxia',inplace=True)
hydropower['State/Province 1'].replace('Guangxi Zhuang Autonomous Region','Guangxi',inplace=True)
hydropower['State/Province 1'].replace('Hubei Province','Hubei',inplace=True)
hydropower['State/Province 1'].replace('Fujiang','Fujian',inplace=True)
hydropower['State/Province 1'].replace('Beijing City','Beijing',inplace=True)
```

省份边界上的进行分配,粗略一省一半

```python
total_Sichuan_Yunnan=hydropower_all[hydropower_all["State/Province 1"]=='Sichuan - Yunnan border']['Capacity (MW)'].sum()
total_Sichuan_Tibet=hydropower_all[hydropower_all["State/Province 1"]=='Sichuan - Tibet border']['Capacity (MW)'].sum()
total_Yunnan_Guizhou=hydropower_all[hydropower_all["State/Province 1"]=='Yunnan - Guizhou border']['Capacity (MW)'].sum()
total_Guangxi_Guizhou=hydropower_all[hydropower_all["State/Province 1"]=='Guangxi - Guizhou border']['Capacity (MW)'].sum()
total_Sichuan_Qinghai=hydropower_all[hydropower_all["State/Province 1"]=='Sichuan - Qinghai border']['Capacity (MW)'].sum()


```

```python
hydropower_capacity= hydropower.groupby("State/Province 1")["Capacity (MW)"].sum()
hydropower_capacity=hydropower_capacity.drop(["Sichuan - Yunnan border","Sichuan - Tibet border","Yunnan - Guizhou border","Guangxi - Guizhou border","Sichuan - Qinghai border"])
hydropower_capacity
```

```python
hydropower_capacity.loc['Sichuan']=hydropower_capacity.loc['Sichuan']+total_Sichuan_Yunnan*0.5+total_Sichuan_Tibet*0.5+total_Sichuan_Qinghai*0.5
hydropower_capacity.loc['Yunnan']+=total_Sichuan_Yunnan*0.5+total_Yunnan_Guizhou*0.5
hydropower_capacity.loc['Tibet']+=total_Sichuan_Tibet*0.5
hydropower_capacity.loc['Guizhou']+=total_Yunnan_Guizhou*0.5+total_Guangxi_Guizhou*0.5
hydropower_capacity.loc['Guangxi']+=total_Guangxi_Guizhou*0.5
hydropower_capacity.loc['Qinghai']+=total_Sichuan_Qinghai*0.5

hours=sum([3579,3613,3726,3827,3622,3412])/6  #http://www.hydropower.org.cn/showNewsDetail.asp?nsId=31457
hydropower_potential=hydropower_capacity*hours
hydropower_potential.name="hydro"
hydropower_potential
```

```python
hydropower_potential.loc['Taiwan']=hydropower_potential.loc['Fujian']*3.59/12.14 #台湾与福建面积之比
  #单位：兆瓦时（e3KWh）
#转变为吉瓦时（e9KWh）
hydropower_potential=hydropower_potential/1e6
```

```python
hydropower_potential
```

```python
hydropower_potential.to_csv("../data/renewable potential/hydropower_potential.csv")
```

整理所有可再生能源的发电量到一张表中

```python
W_hydro=pd.read_csv("../data/renewable potential/hydropower_potential.csv",index_col=0)
W_hydro.index.name="province"
W_nuclear=pd.read_csv("../data/renewable potential/nuclear_potential.csv",index_col=0)
W_nuclear.index.name="province"
W_wind_solar=pd.read_csv("../data/renewable potential/wind_solar_potential.csv",index_col=0)
W_wind_solar.index.name="province"
```

```python
W_wind_solar
```

```python
W=W_wind_solar.copy()
for i in list(W_hydro.index):
    W.loc[i,"W_hydro"]=W_hydro.loc[i]["hydro"]
for j in list(W_nuclear.index):
    W.loc[j,"W_nuclear"]=W_nuclear.loc[j]["nuclear"]
W
```

```python
W.fillna(0,inplace=True)
```

```python
W["W_total"]=W["W_hydro"]+W["W_nuclear"]+W["W_Wind"]+W["W_PV"]
W.to_csv("../data/renewable potential/total_potential.csv")
```
