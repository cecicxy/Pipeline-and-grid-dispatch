# HHI :/

```python
def HHI(G, cable):
    device=pd.DataFrame(index=cable.index,columns=['device'])
    for i in cable.index:
        s=sum(cable.loc[:,i])+G.loc[i,'G']
        device.loc[i,'device']=s

    HHI=sum([(i/sum(device['device']))**2 for i in device['device']])
    return HHI
y4=[]
for w in [0,0.25,0.5,0.75,1]:
    G=pd.read_csv(f'../publication/Figure/dispatch/dispatch_new/year2060_w{w}_cable/year2060_w{w}_cable_G.csv',index_col=0)
    cable=pd.read_csv(f'../publication/Figure/dispatch/dispatch_new/year2060_w{w}_cable/year2060_w{w}_cable_cable.csv',index_col=0)
    y4.append(HHI(G,cable))
fig, ax1 = plt.subplots()

#  y 
ax1.set_xlabel(r'Balance coefficient $\omega$')
ax1.set_ylabel(r'HHI')
ax1.plot(x, y4, marker='o',color='black')
ax1.tick_params(axis='y')

plt.savefig('../publication/Figure/dispatch/new/HHI_index.pdf', format='pdf')
plt.show()


    
```
